```python
import numpy as np
import tensorflow as tf


text = "hello this is a simple handwritten text generation project "

chars = sorted(list(set(text)))
char2idx = {c:i for i,c in enumerate(chars)}
idx2char = {i:c for i,c in enumerate(chars)}


data = np.array([char2idx[c] for c in text])


seq_length = 10
X = []
y = []

for i in range(len(data)-seq_length):
    X.append(data[i:i+seq_length])
    y.append(data[i+seq_length])

X = np.array(X)
y = np.array(y)


model = tf.keras.Sequential([
    tf.keras.layers.Embedding(len(chars), 16),
    tf.keras.layers.LSTM(32),
    tf.keras.layers.Dense(len(chars), activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')

model.fit(X, y, epochs=30)


input_seq = X[0]

for _ in range(50):
    pred = model.predict(input_seq.reshape(1,-1))
    next_char = np.random.choice(len(chars), p=pred[0])
    print(idx2char[next_char], end='')
    input_seq = np.append(input_seq[1:], next_char)
```

```python
input_seq = X[0]

generated = ""

for _ in range(100):
    pred = model.predict(input_seq.reshape(1,-1), verbose=0)
    next_char = np.random.choice(len(chars), p=pred[0])

    generated += idx2char[next_char]

    input_seq = np.append(input_seq[1:], next_char)

print("Generated Text:\n", generated)
```
