```python
import zipfile
import os

with zipfile.ZipFile('/content/archive (3).zip', 'r') as zip_ref:
    zip_ref.extractall('/content')

print(os.listdir('/content'))

import pandas as pd

df = pd.read_csv('/content/spam.csv', encoding='latin-1')
df = df[['v1','v2']]
df.columns = ['label','message']


df['label'] = df['label'].map({'ham':0, 'spam':1})


from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(
    df['message'], df['label'], test_size=0.2, random_state=42
)

vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)

model = MultinomialNB()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))

msg = ["Congratulations! You won a free gift"]
msg_vec = vectorizer.transform(msg)
print("Prediction:", model.predict(msg_vec))
```

```python
msgs = [
    "Win a free iPhone now!!!",
    "Hey, are you done with preparing for your presentation?",
    "Congratulations! You have won cash prize",
    "Let's meet at the cafe."
]

msgs_vec = vectorizer.transform(msgs)
preds = model.predict(msgs_vec)

for i in range(len(msgs)):
    print(msgs[i])
    print("Prediction:", "Spam" if preds[i]==1 else "Ham")
    print()
```
